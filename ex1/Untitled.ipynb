```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ex1data2.txt', delimiter = ',')
m = np.shape(data)[0]
n = np.shape(data)[1]
X = np.reshape(data[:,0:n-1], [m,n-1])
y = np.reshape(data[:,n-1], [m,1])


f, (ax1, ax2) = plt.subplots(2)
ax1.plot(X[:,0], y, 'ro')
ax2.scatter(X[:,1], y)
plt.show()
```

```python
from numpy.linalg import inv

def normal_equations(X,y):
    X = np.column_stack([np.ones([m,1]), X])
    inverted = inv(np.matmul(X.T,X))
    theta = np.matmul(np.matmul(inverted,X.T),y)
    return theta
```

```python
theta = normal_equations(X,y)
theta
```
